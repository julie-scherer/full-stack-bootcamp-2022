### Exercise 1 - Turn the shopping cart program into an object-oriented program

See procedure-orientated program here: https://github.com/schererjulie/coding_temple/blob/main/week2/shopping_cart/shopping_cart_program.ipynb

```python
class Plurality():
    def __init__(self, item='') -> None:
        self.item = item

    def singular(self):
        if self.item[-1] != 's':
            pass
        
        elif self.item[-3] == 'i' and self.item[-2] == 'e' and self.item-[1] == 's':
            self.item = self.item.rstrip('ies') + 'y'
        elif self.item[-3] == 'o' and self.item[-2] == 'e' and self.item[-1] == 's':
            self.item = self.item.rstrip('es')
        elif self.item[-1] == 's':
            self.item = self.item.rstrip('s')
        return self.item

    def plural(self):
        if self.item[-1] == 's':
            pass
        elif self.item[-3] == 'a' and self.item[-2] == 't' and self.item[-1] == 'o':
            self.item = self.item + 'es'
        elif self.item[-1] == 'y':
            self.item = self.item.rstrip('y') + 'ies'
        elif self.item[-2] == 'c' and self.item[-1] == 'h':
            self.item = self.item + 'es'
        else:
            self.item = self.item + 's'
        return self.item



class Cart():
    def __init__(self, shop_list) -> None:
        self.shop_list = shop_list
    
    
    ### INITALIZE PROGRAM
    def initalize(self):
        action = input("What would you like to do? add/remove/view/clear/quit  ").lower()
        if ((action == 'add') or (action == 'remove') or (action == 'view') or (action == 'clear') or (action == 'quit')) == False:
            action = input("This is an invalid response. Please enter add, remove, view, clear, or quit. ")
        return action


    ### ADD ITEM TO SHOP LIST
    def add_item(self):
        
        while True:
            item_input = input("What would you like to add to your cart? ").lower()
            char_list = [char.isdigit() for char in item_input]
            if True not in char_list:
                item = Plurality(item_input).singular().lower()
                break
            else:
                print("Invalid response.")
        
        while True:
            no_input = input(f"Please enter the number of {Plurality(item).plural()} you would like to add. ")
            try: # ~ try converting to int type
                no_items = int(no_input)
                break
            except:
                print("Invalid response.")
        
        while True:
            cost_input = input(f"Please enter the cost per {Plurality(item).singular()} in $0.00 format. ").strip('$')
            try:
                total_cost = round(float(cost_input) * no_items, 2) # ~ convert total cost (no of items * cost per item) to float w/ 2 decimals
                break
            except:
                print("Invalid response.")
        
        self.shop_list[item] = {'No':no_items, 'Cost':total_cost}
        print(f"{no_items}x {item.capitalize()} added to your cart.")
        return self.shop_list


    ### REMOVE ITEM FROM SHOP LIST
    def remove_item(self):
        
        while True:
            rm_item_input = input("What would you like to remove from your cart? Please type the item name in singular format. ").lower()
            char_list2 = [char.isdigit() for char in rm_item_input]
            if True not in char_list2:
                rm_item = Plurality(rm_item_input).singular().lower()
                break
            else:
                print("Invalid response.")
        
        while True:
            if self.shop_list.get(rm_item_input) == None:
                rm_item_input2 = input(f"You do not have '{rm_item_input}' in your cart. Please enter an existing item in your cart (in singular format) or type 'exit' to return to the main menu. ").lower()  # ~ if item is NOT in shop_list, print error message and ask user to re-enter the item or exit to main menu
                if rm_item_input2 == 'exit': # if the user enters 'exit', exit the function
                    return self.shop_list
                else:
                    rm_item_input = rm_item_input2
            else:
                break
        
        while True:
            rm_no_input = input(f"Please enter the number of {Plurality(rm_item).plural()} you would like to remove. ")
            try: # ~ try converting to int type
                num_to_remove = int(rm_no_input)
                break
            except:
                print("Invalid response.")
        
        if num_to_remove == 0:
            print(f"No {rm_items} have been removed from your cart.")
            return self.shop_list
        
        cur_item_quant = self.shop_list[rm_item]['No'] # ~ save the current number of items in shop_list
        print("CURRENT QUANTITY ",cur_item_quant)
        
        # ~ if user enters a value greater than the number of available items
        while num_to_remove > cur_item_quant:
            num_error = input(f"You only have {cur_item_quant} {Plurality(rm_item).plural() if cur_item_quant > 1 else Plurality(rm_item).singular()} in your cart. Please re-enter the number of items you would like to remove or type 'exit' to return to the main menu. ")
            if num_error == 'exit':
                return self.shop_list
            elif num_error.isdigit():
                num_to_remove = int(num_error)
            else:
                print("Invalid response.")
        
        # ~ if user enters a value that is the same as the current item quantity...
        while True:
            if cur_item_quant == num_to_remove: # or cur_item_quant - new_item_quant == 0
                confirm = input(f"{rm_item.capitalize()} will be deleted from your cart. Are you sure you want to continue? y/n ").lower() # ~ alert the user that the item will be deleted from their cart and ask if they want to continue
                if confirm == 'y': # ~ if the user enters 'y', delete the item
                    del self.shop_list[rm_item]
                    print(f"1 {Plurality(rm_item).singular()} has been removed from your cart.") if cur_item_quant == 1 else print(f"{cur_item_quant} {Plurality(rm_item).plural()} have been removed from your cart.")
                    return self.shop_list
                elif confirm == 'n': # ~ if the user enters 'n', ask if they would like to 'restart' (return to the beginning of the method) or 'exit' (return to main menu)
                    print('Returning to main menu...')
                    return self.shop_list
                else:
                    print("Invalid response.")
        
        # once num_to_remove is fainlized, calculate the new item quantity, or difference between the current number of items (item quantity) and number of items the user would like to remove
        new_item_quant = cur_item_quant - num_to_remove 
        
        old_cost = self.shop_list.get(rm_item).get('Cost')
        new_cost = (old_cost / cur_item_quant) * new_item_quant
        self.shop_list[rm_item] = {'Cost':new_cost, 'No':new_item_quant}
        print(f"You now have 1 {Plurality(rm_item).singular()} in your cart.") if new_item_quant == 1 else print(f"You now have {new_item_quant} {Plurality(rm_item).plural()} in your cart.")
        
        return self.shop_list
    


    ### VIEW/DISPLAY CART
    def view_cart(self):
        if len(self.shop_list) == 0:
            print("There are no items in your cart.")
        else:
            total_cost = 0
            print("\nYour current cart:")
            for key, value in self.shop_list.items():
                no_items = value['No']
                cost = value['Cost']
                total_cost += cost
                if no_items == 1:
                    print("{}x {} (${:.2f})".format(no_items, Plurality(key).singular().title(),cost))
                else:
                    print("{}x {} (${:.2f})".format(no_items, Plurality(key).singular().title(),cost))
            tax = total_cost * .08
            print("-------------------")
            print("Subtotal = ${:.2f}".format(total_cost))
            print("Tax      = ${:.2f}".format(tax))
            print("Total    = ${:.2f}\n".format(total_cost + tax))
        return self.shop_list
    


    ### CLEAR CART
    def clear_cart(self):
        self.shop_list = {}
        return shop_list
    

    
    ### QUIT PROGRAM
    def quit(self):
        print("Thank you for shopping with us!")


##########################################################
### RUN ################################################## 
##########################################################

def main():
    cart1 = Cart(shop_list={})
    cart1.add_item()
    while True:
        action = cart1.initalize()
        if action == "add":
            cart1.add_item()
        elif action == "remove":
            cart1.remove_item()
        elif action == "view":
            cart1.view_cart()
        elif action == "clear":
            cart1.clear_cart()
        elif action == "quit":
            cart1.quit()
            break

main()
```
