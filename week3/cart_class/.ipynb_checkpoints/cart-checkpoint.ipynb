### Exercise 1 - Turn the shopping cart program into an object-oriented program

See procedure-orientated program here: https://github.com/schererjulie/coding_temple/blob/main/week2/shopping_cart/shopping_cart_program.ipynb

```python
class Plurality():
    def __init__(self, item='') -> None:
        self.item = item

    def singular(self):
        if self.item[-1] != 's':
            pass
        
        elif self.item[-3] == 'i' and self.item[-2] == 'e' and self.item-[1] == 's':
            self.item = self.item.rstrip('ies') + 'y'
        elif self.item[-3] == 'o' and self.item[-2] == 'e' and self.item[-1] == 's':
            self.item = self.item.rstrip('es')
        elif self.item[-1] == 's':
            self.item = self.item.rstrip('s')
        return self.item

    def plural(self):
        if self.item[-1] == 's':
            pass
        elif self.item[-3] == 'a' and self.item[-2] == 't' and self.item[-1] == 'o':
            self.item = self.item + 'es'
        elif self.item[-1] == 'y':
            self.item = self.item.rstrip('y') + 'ies'
        elif self.item[-2] == 'c' and self.item[-1] == 'h':
            self.item = self.item + 'es'
        else:
            self.item = self.item + 's'
        return self.item

import time
class Cart():    
    spec_char_list = list('[@_!#$%^&*()<>?/\|}{~:]')
    num_list = list('0123456789')

    def __init__(self, shop_list) -> None:
        self.shop_list = shop_list
    
    ### INITALIZE PROGRAM
    # ---------------------------------------------------------------
    def initalize(self):
        action = input("What would you like to do? add/remove/view/clear/quit  ").lower()
        if ((action == 'add') or (action == 'remove') or (action == 'view') or (action == 'clear') or (action == 'quit')) == False:
            action = input("This is an invalid response. Please enter add, remove, view, clear, or quit. ")
        return action

    ### ADD ITEM TO SHOP LIST
    # ---------------------------------------------------------------
    def add_item(self):
        while True:
            item_input = input("What would you like to add to your cart? ").lower()
            for char in item_input:
                if (char in Cart.spec_char_list) or (char in Cart.num_list):
                    print("Invalid response.")
                    break
            else:
                break
        item = Plurality(item_input).singular().lower()
        print(f"Item selected: {item.capitalize()}")
        
        while True:
            item_quant_input = input(f"Please enter the number of {Plurality(item).plural()} you would like to add to your cart. ")
            if item_quant_input.isdigit():
                break
            else:
                print("Invalid response.")
        item_quant = int(item_quant_input)
        print(f"{item_quant}x {item.capitalize()} added to your cart.")
        
        while True:
            cost_input = input(f"Please enter the cost per {Plurality(item).singular()} in $0.00 format. ").strip('$')
            try:
                float(cost_input)
                break
            except:
                print("Invalid response.")
        cost_per_item = cost_input
        total_cost = round(float(cost_per_item) * item_quant, 2) # ~ convert total cost (no of items * cost per item) to float
        print(f"Total cost: ${total_cost}")
        self.shop_list[item] = {
            'Quant': item_quant, 
            'Cost': total_cost
            }
        return self.shop_list
        # Example: {'pickle': {'Quant': 5, 'Cost': 5.0}}
    
    ### REMOVE ITEM FROM SHOP LIST
    # ---------------------------------------------------------------
    def remove_item(self):
        
        print(f"Items currently in your shopping cart: {[key for key in self.shop_list]}")
#         print(self.shop_list)
        item_input = input("What item would you like to remove from your cart? ").lower()
        for char in item_input:
            while (char in Cart.spec_char_list) or (char in Cart.num_list):
                print("Invalid response. Responses cannot include special characters or numbers. Please try again.")
                item_input = input("What would you like to remove from your cart? ").lower()
        item = Plurality(item_input).singular().lower()
        
        while item not in self.shop_list.keys():
            reselect_item = input(f"You do not have '{item_input}' in your cart. Please enter an existing item in your cart (in singular format) or type 'exit' to return to the main menu. ").lower()  # ~ if item is NOT in shop_list, print error message and ask user to re-enter the item or exit to main menu
            if reselect_item == 'exit':
                return self.shop_list
            else:
                item = Plurality(reselect_item).singular().lower()
        print(f"Item selected: {item.capitalize()}")
        
        while True:
            item_quant_input = input(f"How many {Plurality(item).plural()} would you like to remove from your cart? Enter 0 or type 'exit' if you no longer wish to make any changes. ")
            if item_quant_input == '0' or item_quant_input == 'exit':
                print(f"Returning to the main menu...")
                time.sleep(1)
                return self.shop_list
            if item_quant_input.isdigit():
                break
            else:
                print("Invalid response. Please try again.")
        rm_item_quant = int(item_quant_input)

        cur_item_quant = self.shop_list[item]['Quant'] # ~ save the current number of items in shop_list
        
        while rm_item_quant > cur_item_quant: # ~ if the user tries to remove more than what's available in their cart 
            num_error = input(f"You only have {cur_item_quant} {Plurality(item).plural() if cur_item_quant > 1 else Plurality(item).singular()} in your cart. Please re-enter the number of {Plurality(item).plural()} you want to remove from your cart, or type 'exit' to return to the main menu. ")
            if num_error == 'exit':
                print(f"Returning to the main menu...")
                time.sleep(1)
                return self.shop_list
            if num_error.isdigit():
                rm_item_quant = int(num_error)
                break
            else:
                print("Invalid response.")
        
        if cur_item_quant == rm_item_quant: # ~ if user wants to remove all quantities... also, cur_item_quant - new_item_quant == 0
            while True:
                confirm = input(f"{item.capitalize()} will be deleted from your cart. Are you sure you want to continue? y/n ").lower() # ~ alert the user that the item will be deleted and ask if they want to continue
                if confirm == 'y': # ~ if the user enters 'y', delete the item
                    del self.shop_list[item]
                    print(f"Removing {Plurality(item).singular()} from your cart...") if rm_item_quant == 1 else print(f"Removing all {Plurality(item).plural()} from your cart...")
                    time.sleep(1)
                    return self.shop_list
                elif confirm == 'n': # ~ if the user enters 'n', return to the main menu
                    print('Returning to main menu...')
                    return self.shop_list
                else:
                    print("Invalid response. Please try again.")
                    continue
    
        # calculate the new item quantity, or difference between the current number of items (item quantity) and number of items the user would like to remove
        new_item_quant = cur_item_quant - rm_item_quant 
        cur_total = self.shop_list.get(item).get('Cost')
        new_cost = (cur_total / cur_item_quant) * new_item_quant
        self.shop_list[item] = {
            'Cost': new_cost, 
            'Quant': new_item_quant
            }
        print(f"You now have 1 {Plurality(item).singular()} in your cart.") if new_item_quant == 1 else print(f"You now have {new_item_quant} {Plurality(item).plural()} in your cart.")
        
        return self.shop_list

    ### VIEW/DISPLAY CART
    # ---------------------------------------------------------------
    def view_cart(self):
        if len(self.shop_list) == 0:
            print("There are no items in your cart.")
        else:
            total_cost = 0
            print("\nYour current cart:")
            for key, value in self.shop_list.items():
                no_items = value['Quant']
                cost = value['Cost']
                total_cost += cost
                if no_items == 1:
                    print("{}x {} (${:.2f})".format(no_items, Plurality(key).singular().title(),cost))
                else:
                    print("{}x {} (${:.2f})".format(no_items, Plurality(key).singular().title(),cost))
            tax = total_cost * .08
            print("-------------------")
            print("Subtotal = ${:.2f}".format(total_cost))
            print("Tax      = ${:.2f}".format(tax))
            print("Total    = ${:.2f}\n".format(total_cost + tax))
        return self.shop_list
    
    ### CLEAR CART
    # ---------------------------------------------------------------
    def clear_cart(self):
        self.shop_list = {}
        print('Removing all items from your cart...')
        time.sleep(1)
        return self.shop_list
    
    ### QUIT PROGRAM
    # ---------------------------------------------------------------
    def quit(self):
        print("\nThank you for shopping with us!\n")


##########################################################
### RUN ################################################## 
##########################################################

def main():
    cart1 = Cart(shop_list={})
    cart1.add_item()
    while True:
        action = cart1.initalize()
        if action == "add":
            cart1.add_item()
        elif action == "remove":
            cart1.remove_item()
        elif action == "view":
            cart1.view_cart()
        elif action == "clear":
            cart1.clear_cart()
        elif action == "quit":
            cart1.quit()
            break

main()
```
